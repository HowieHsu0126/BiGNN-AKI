```python
import dask.dataframe as dd

aki_eicu = dd.read_csv('csv/aki_eicu.csv')
```

```python
unique_count = aki_eicu['patientunitstayid'].nunique().compute()
unique_aki_count = aki_eicu[aki_eicu['final_aki_status']!='No ICU Acquired AKI']['patientunitstayid'].nunique().compute()
print(f"总共有 {unique_count} 名病人，其中有 {unique_aki_count} 患急性肾损伤（AKI）的病人")
```

```python
# Filter the DataFrame for 'ICU Acquired AKI' and select the 'patientunitstayid' column
icu_aki_ids_df = aki_eicu[aki_eicu['final_aki_status'] == 'ICU Acquired AKI']['patientunitstayid'].compute()

# Save the resulting IDs to a CSV file
output_path = 'csv/aki_eicu_id.csv'
icu_aki_ids_df.to_csv(output_path, index=False)
```
